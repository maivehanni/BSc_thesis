```python
#Lab data
import matplotlib.pyplot as plt
import pandas as pd

GR = [0.049, 0.100, 0.151, 0.203, 0.301]
glc = [0.476, 1.114, 1.648, 2.305, 3.1]
co2 = [1.171, 2.521, 3.854, 5.834, 7.415]
o2 = [1.083, 2.521, 3.851, 4.352, 6.327]

data = {'GR': GR, 'glc': glc, 'o2': o2, 'co2': co2}

lab_data = pd.DataFrame(data)


names = ['D-glucose exchange', 'Oxygen exchange', 'Carbon dioxide exchange']



# 'Biomass growth rate'

x = lab_data['GR']
y = lab_data[['glc', 'o2', 'co2']]

plt.plot(x, y, '-', label= names) #
plt.legend(fontsize=6, loc='upper left')
plt.title("Experimental exchange fluxes") 
plt.xlabel('Biomass growth rate $(1/h)$')
plt.ylabel('Flux $(g/gDW/h)$')
```
