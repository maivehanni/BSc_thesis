```python
import cobra
import pandas as pd
import os
from os.path import join
from cobra import Model, Reaction, Metabolite
from cobra.sampling import sampling
import numpy as np
os.environ["R_HOME"] = f"{os.environ['CONDA_PREFIX']}\\Lib\\R"
import rpy2.robjects
from plotnine import *
import matplotlib.pyplot as plt

model=cobra.io.read_sbml_model("C:\\Users\\Maive\\Desktop\\BSc_loputoo\\Model_files\\Rt_IFO0880.xml")
model.objective = "BIOMASS_RT" 
```

```python
# Get nadph fluxes at lowest biomass growth rate
nadphs = ['nadph_c', 'nadph_m', 'nadph_x']

nadph_fluxes_lowestGR = pd.DataFrame() #model.metabolites.nadph_c.summary().producing_flux 

medium = model.medium
medium["EX_glc__D_e"] = glucose_uptakes[0]
model.medium = medium
    
solution = model.optimize()
biomass_GR_lowest = solution.objective_value
for metabolite in nadphs:
    nadph_fluxes_lowestGR = pd.concat([nadph_fluxes_lowestGR, getattr(model.metabolites, metabolite).summary().consuming_flux, getattr(model.metabolites, metabolite).summary().producing_flux])

nadph_fluxes_lowestGR = nadph_fluxes_lowestGR.sort_values(by='flux', ascending=False).drop(columns = ['percent'])
nadph_fluxes_lowestGR = nadph_fluxes_lowestGR[(nadph_fluxes_lowestGR['flux']) != 0.0] # for getting non-zero fluxes only
nadph_sum_producing_flux = sum(nadph_fluxes_lowestGR[nadph_fluxes_lowestGR['flux'] > 0]['flux'])
nadph_sum_consuming_flux = sum(nadph_fluxes_lowestGR[nadph_fluxes_lowestGR['flux'] < 0]['flux'])
print(f'SUM produced: {nadph_sum_producing_flux}, SUM consumed: {nadph_sum_consuming_flux}')

nadph_fluxes_lowestGR
```

```python
# Get nadph fluxes at highest biomass growth rate
nadph_fluxes_highestGR = pd.DataFrame() #model.metabolites.nadph_c.summary().producing_flux 

medium = model.medium
medium["EX_glc__D_e"] = glucose_uptakes[5]
model.medium = medium
    
solution = model.optimize()
biomass_GR_highest = solution.objective_value
for metabolite in nadphs:
    nadph_fluxes_highestGR = pd.concat([nadph_fluxes_highestGR, getattr(model.metabolites, metabolite).summary().consuming_flux, getattr(model.metabolites, metabolite).summary().producing_flux])

nadph_fluxes_highestGR = nadph_fluxes_highestGR.sort_values(by='flux', ascending=False).drop(columns = ['percent'])
nadph_fluxes_highestGR = nadph_fluxes_highestGR[(nadph_fluxes_highestGR['flux']) != 0.0] # for getting non-zero fluxes only
nadph_fluxes_highestGR
```
